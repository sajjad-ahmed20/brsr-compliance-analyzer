# BRSR Compliance Analyzer
## Tata Motors Limited (TML) | FY 2024-25
### Principles Analyzed: P1 (Ethics), P6 (Environment), P8 (Inclusive Growth)

**Methodology:**
Each indicator is scored across three dimensions:
- **Completeness (40 pts):** Was the indicator disclosed at all?
- **Quality (35 pts):** Is the disclosure quantitative and specific?
- **Verification (25 pts):** Was it externally assured (e.g. KPMG)?

**Data Source:** Tata Motors BRSR FY2024-25 (80th Integrated Annual Report)

```python
def score_principle(principle_data):
    all_indicators = principle_data["essential"] + principle_data["leadership"]
    total = len(all_indicators)

    disclosed = sum(1 for i in all_indicators if i["disclosed"])
    quantitative = sum(1 for i in all_indicators if i["quantitative"])
    verified = sum(1 for i in all_indicators if i["verified"])
    gaps = [i for i in all_indicators if i.get("gap")]

    completeness = round((disclosed / total) * 40, 1)
    quality = round((quantitative / total) * 35, 1)
    verification = round((verified / total) * 25, 1)
    total_score = round(completeness + quality + verification, 1)

    if total_score >= 80:
        verdict = "STRONG"
        verdict_color = "#27AE60"
    elif total_score >= 60:
        verdict = "ADEQUATE"
        verdict_color = "#F39C12"
    else:
        verdict = "NEEDS IMPROVEMENT"
        verdict_color = "#E74C3C"

    return {
        "total": total,
        "essential_count": len(principle_data["essential"]),
        "leadership_count": len(principle_data["leadership"]),
        "disclosed": disclosed,
        "quantitative": quantitative,
        "verified": verified,
        "completeness": completeness,
        "quality": quality,
        "verification": verification,
        "total_score": total_score,
        "verdict": verdict,
        "verdict_color": verdict_color,
        "gaps": gaps,
        "name": principle_data["name"],
        "color": principle_data["color"]
    }

results = {p_id: score_principle(p_data) for p_id, p_data in indicators.items()}

print("=" * 65)
print("  BRSR COMPLIANCE SCORECARD — TATA MOTORS FY2024-25")
print("=" * 65)
for p_id, r in results.items():
    print(f"\n  {p_id} — {r['name']}")
    print(f"  {'─'*55}")
    print(f"  Completeness  : {r['disclosed']}/{r['total']} disclosed     → {r['completeness']:>5}/40")
    print(f"  Quality       : {r['quantitative']}/{r['total']} quantitative  → {r['quality']:>5}/35")
    print(f"  Verification  : {r['verified']}/{r['total']} assured       → {r['verification']:>5}/25")
    print(f"  TOTAL SCORE   : {r['total_score']}/100  [{r['verdict']}]")
    if r['gaps']:
        print(f"  Gaps identified: {len(r['gaps'])}")
        for g in r['gaps']:
            print(f"    → {g['indicator']}: {g['gap']}")
```

```python
DARK = '#1A1A2E'
CARD = '#16213E'
WHITE = '#FFFFFF'
MUTED = '#BDC3C7'

fig = plt.figure(figsize=(22, 16), facecolor=DARK)
gs = gridspec.GridSpec(4, 3, figure=fig,
                        hspace=0.50, wspace=0.30,
                        left=0.04, right=0.97,
                        top=0.91, bottom=0.04)

# Title
fig.text(0.5, 0.96, 'BRSR COMPLIANCE INTELLIGENCE REPORT',
         ha='center', fontsize=20, fontweight='bold', color=WHITE)
fig.text(0.5, 0.935,
         'Tata Motors Limited  |  FY 2024-25  |  Principles: P1 Ethics · P6 Environment · P8 Inclusive Growth  |  Source: 80th Integrated Annual Report',
         ha='center', fontsize=10, color=MUTED)

# ── ROW 1: Score KPI cards ──────────────────────────────────
kpis = [
    ("P1 — Ethics", results["P1"]["total_score"], results["P1"]["verdict"], results["P1"]["color"]),
    ("P6 — Environment", results["P6"]["total_score"], results["P6"]["verdict"], results["P6"]["color"]),
    ("P8 — Inclusive Growth", results["P8"]["total_score"], results["P8"]["verdict"], results["P8"]["color"]),
]
overall = round(sum(r["total_score"] for r in results.values()) / 3, 1)

for i, (label, score, verdict, color) in enumerate(kpis):
    ax = fig.add_subplot(gs[0, i])
    ax.set_facecolor(CARD)
    ax.set_xlim(0, 1); ax.set_ylim(0, 1); ax.axis('off')
    ax.add_patch(FancyBboxPatch((0.03, 0.03), 0.94, 0.94,
                                 boxstyle='round,pad=0.02',
                                 facecolor=CARD, edgecolor=color,
                                 linewidth=2.5, transform=ax.transAxes))
    ax.text(0.5, 0.78, label, ha='center', fontsize=9,
            color=MUTED, fontweight='bold', transform=ax.transAxes)
    ax.text(0.5, 0.48, f"{score}", ha='center', fontsize=26,
            color=color, fontweight='bold', transform=ax.transAxes)
    ax.text(0.5, 0.28, "/ 100", ha='center', fontsize=10,
            color='#7F8C8D', transform=ax.transAxes)
    ax.text(0.5, 0.12, verdict, ha='center', fontsize=8,
            color=color, fontweight='bold', transform=ax.transAxes)

# ── ROW 2 LEFT: Radar / Dimension breakdown bars ────────────
ax_bars = fig.add_subplot(gs[1, 0:2])
ax_bars.set_facecolor(DARK)

principles = list(results.keys())
x = np.arange(len(principles))
width = 0.25

comp_vals = [r["completeness"] for r in results.values()]
qual_vals = [r["quality"] for r in results.values()]
ver_vals = [r["verification"] for r in results.values()]

b1 = ax_bars.bar(x - width, comp_vals, width, label='Completeness (/40)',
                  color='#3498DB', edgecolor='none')
b2 = ax_bars.bar(x, qual_vals, width, label='Quality (/35)',
                  color='#9B59B6', edgecolor='none')
b3 = ax_bars.bar(x + width, ver_vals, width, label='Verification (/25)',
                  color='#1ABC9C', edgecolor='none')

for bars in [b1, b2, b3]:
    for bar in bars:
        h = bar.get_height()
        ax_bars.text(bar.get_x() + bar.get_width()/2., h + 0.4,
                     f'{h:.0f}', ha='center', va='bottom',
                     color=WHITE, fontsize=8, fontweight='bold')

ax_bars.set_facecolor(DARK)
ax_bars.set_xticks(x)
ax_bars.set_xticklabels([f"{p}\n{results[p]['name']}" for p in principles],
                         color=WHITE, fontsize=9)
ax_bars.tick_params(colors=WHITE)
ax_bars.set_ylabel('Score', color=MUTED, fontsize=9)
ax_bars.set_ylim(0, 48)
ax_bars.spines['bottom'].set_color('#444')
ax_bars.spines['left'].set_color('#444')
ax_bars.legend(fontsize=8, frameon=False, labelcolor=WHITE,
               loc='upper right')
ax_bars.set_title('Score Breakdown by Dimension', color=WHITE,
                   fontsize=12, fontweight='bold', pad=12)
ax_bars.grid(axis='y', alpha=0.1, color=WHITE)

# ── ROW 2 RIGHT: Overall score donut ───────────────────────
ax_donut = fig.add_subplot(gs[1, 2])
ax_donut.set_facecolor(DARK)

donut_vals = [r["total_score"] for r in results.values()]
donut_colors = [r["color"] for r in results.values()]

wedges, texts, autotexts = ax_donut.pie(
    donut_vals, colors=donut_colors,
    autopct='%1.0f%%', pctdistance=0.82,
    wedgeprops=dict(width=0.55, edgecolor=DARK, linewidth=2),
    startangle=90
)
for at in autotexts:
    at.set_color(WHITE); at.set_fontsize(9); at.set_fontweight('bold')

centre = plt.Circle((0, 0), 0.45, fc=DARK)
ax_donut.add_patch(centre)
ax_donut.text(0, 0.1, f'{overall}', ha='center', fontsize=18,
               color=WHITE, fontweight='bold')
ax_donut.text(0, -0.15, 'Avg Score', ha='center', fontsize=8, color=MUTED)
ax_donut.legend(['P1 Ethics', 'P6 Environment', 'P8 Inclusive Growth'],
                 loc='lower center', ncol=3, fontsize=8,
                 frameon=False, labelcolor=WHITE,
                 bbox_to_anchor=(0.5, -0.08))
ax_donut.set_title('Overall Compliance Score', color=WHITE,
                    fontsize=12, fontweight='bold', pad=12)

# ── ROW 3: Disclosure heatmap ───────────────────────────────
ax_heat = fig.add_subplot(gs[2, 0:2])
ax_heat.set_facecolor(DARK)

all_inds = []
for p_id, p_data in indicators.items():
    for ind in p_data["essential"] + p_data["leadership"]:
        all_inds.append({
            "principle": p_id,
            "id": ind["id"],
            "short": ind["indicator"][:38] + ("…" if len(ind["indicator"]) > 38 else ""),
            "disclosed": 1 if ind["disclosed"] else 0,
            "quantitative": 1 if ind["quantitative"] else 0,
            "verified": 1 if ind.get("verified") else 0,
            "type": "E" if "_E" in ind["id"] else "L"
        })

df_heat = pd.DataFrame(all_inds)
heat_matrix = df_heat[["disclosed", "quantitative", "verified"]].values.T

colors_heat = np.where(heat_matrix == 1, 1.0, 0.0)
cmap_custom = plt.cm.colors.LinearSegmentedColormap.from_list(
    'custom', ['#2C3E50', '#27AE60'])

im = ax_heat.imshow(colors_heat, aspect='auto', cmap=cmap_custom,
                     vmin=0, vmax=1, interpolation='nearest')

ax_heat.set_yticks([0, 1, 2])
ax_heat.set_yticklabels(['Disclosed', 'Quantitative', 'Verified'],
                         color=WHITE, fontsize=9)
ax_heat.set_xticks(range(len(df_heat)))
ax_heat.set_xticklabels(
    [f"{r['principle']}\n{r['type']}{r['id'].split('_')[1][1:]}"
     for _, r in df_heat.iterrows()],
    color=MUTED, fontsize=6.5, rotation=45, ha='right')
ax_heat.tick_params(colors=WHITE)
ax_heat.set_title('Indicator Disclosure Heatmap — All 39 Indicators',
                   color=WHITE, fontsize=12, fontweight='bold', pad=12)

legend_patches = [
    mpatches.Patch(color='#27AE60', label='Disclosed / Met'),
    mpatches.Patch(color='#2C3E50', label='Not Disclosed / Not Met')
]
ax_heat.legend(handles=legend_patches, loc='upper right',
               fontsize=8, frameon=False, labelcolor=WHITE)

# ── ROW 3 RIGHT: Gap analysis table ─────────────────────────
ax_gaps = fig.add_subplot(gs[2, 2])
ax_gaps.set_facecolor(DARK)
ax_gaps.axis('off')

gap_rows = []
for p_id, p_data in indicators.items():
    for ind in p_data["essential"] + p_data["leadership"]:
        if ind.get("gap"):
            gap_rows.append([p_id, ind["indicator"][:28] + "…", ind["gap"][:32] + "…"])

if gap_rows:
    table = ax_gaps.table(
        cellText=gap_rows,
        colLabels=["Principle", "Indicator", "Gap Identified"],
        cellLoc='left', loc='center',
        colWidths=[0.12, 0.42, 0.46]
    )
    table.auto_set_font_size(False)
    table.set_fontsize(7.5)
    table.scale(1, 2.2)
    for j in range(3):
        table[0, j].set_facecolor('#C0392B')
        table[0, j].set_text_props(color=WHITE, fontweight='bold')
    for i in range(1, len(gap_rows) + 1):
        for j in range(3):
            table[i, j].set_facecolor('#1A1A2E' if i % 2 == 0 else '#16213E')
            table[i, j].set_text_props(color=WHITE)

ax_gaps.set_title('Disclosure Gaps Identified', color=WHITE,
                   fontsize=10, fontweight='bold', pad=12)

# ── ROW 4: Summary scorecard table ──────────────────────────
ax_summary = fig.add_subplot(gs[3, :])
ax_summary.set_facecolor(DARK)
ax_summary.axis('off')

summary_rows = []
for p_id, r in results.items():
    summary_rows.append([
        f"{p_id} — {r['name']}",
        f"{r['essential_count']} Essential + {r['leadership_count']} Leadership",
        f"{r['disclosed']}/{r['total']}",
        f"{r['quantitative']}/{r['total']}",
        f"{r['verified']}/{r['total']}",
        f"{r['completeness']}/40",
        f"{r['quality']}/35",
        f"{r['verification']}/25",
        f"{r['total_score']}/100",
        r['verdict']
    ])

summary_table = ax_summary.table(
    cellText=summary_rows,
    colLabels=["Principle", "Indicators", "Disclosed",
               "Quantitative", "Verified", "Completeness",
               "Quality", "Verification", "TOTAL", "Verdict"],
    cellLoc='center', loc='center',
    colWidths=[0.22, 0.14, 0.07, 0.09, 0.07,
               0.09, 0.07, 0.09, 0.07, 0.09]
)
summary_table.auto_set_font_size(False)
summary_table.set_fontsize(8.5)
summary_table.scale(1, 2.5)

for j in range(10):
    summary_table[0, j].set_facecolor('#2C3E50')
    summary_table[0, j].set_text_props(color=WHITE, fontweight='bold')

verdict_colors_map = {"STRONG": "#27AE60", "ADEQUATE": "#F39C12",
                       "NEEDS IMPROVEMENT": "#E74C3C"}
p_colors = [r["color"] for r in results.values()]

for i, (p_id, r) in enumerate(results.items(), 1):
    for j in range(10):
        summary_table[i, j].set_facecolor('#16213E')
        summary_table[i, j].set_text_props(color=WHITE)
    summary_table[i, 8].set_text_props(color=p_colors[i-1], fontweight='bold')
    summary_table[i, 9].set_facecolor(verdict_colors_map[r['verdict']])
    summary_table[i, 9].set_text_props(color=WHITE, fontweight='bold')

ax_summary.set_title('Full Compliance Scorecard', color=WHITE,
                      fontsize=11, fontweight='bold', pad=10)

plt.savefig('BRSR_Compliance_Report_TataMotors_FY25.png',
            dpi=150, bbox_inches='tight', facecolor=DARK)
plt.show()
print("BRSR Compliance Report saved")
```
